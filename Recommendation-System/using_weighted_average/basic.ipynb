```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
credits.head()
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
movies.head()
```

```python
credits_new = credits.rename(index=str , columns={"movie_id" : "id"})
merged_movies = movies.merge(credits_new , on="id")
merged_movies.head()
```

```python
merged_movies.shape
```

```python
merged_movies.info()
```

```python
movies_cleaned = merged_movies.drop(columns=["homepage","production_countries","title_x","title_y","status"])
movies_cleaned.head()
```

```python
movies_cleaned.shape
```


Using Weighted average for each movie's Average Rating

image.png

 Source: http://trailerpark.weebly.com/imdb-rating.html?source=post_page---------------------------

```python
v = movies_cleaned["vote_count"]
R = movies_cleaned["vote_average"]
C = movies_cleaned["vote_average"].mean()
m = movies_cleaned["vote_count"].quantile(0.70)
```

```python
movies_cleaned["weighted_average"] = ((R*v)+(C*m)) / (v+m)
```

```python
movies_cleaned.head()
```

```python
movies_ranking = movies_cleaned.sort_values("weighted_average" , ascending=False)
movies_ranking[["original_title","vote_average","vote_count","weighted_average","popularity"]]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average=movies_ranking.sort_values('weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1=sns.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['original_title'].head(10), data=weight_average)
plt.xlim(4, 10)
plt.title('Best Movies by average votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
import matplotlib.pyplot as plt

weight_average = movies_ranking.sort_values("weighted_average" , ascending=False)
plt.bar(weight_average["original_title"].head(10) , weight_average["vote_average"].head(10))
plt.xlabel("Movie Title")
plt.ylabel("Average Rating")
plt.show()
```

```python
popularity=movies_ranking.sort_values('popularity',ascending=False)
plt.figure(figsize=(12,6))
ax=sns.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)

plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_popular_movies.png')
```

```python
popularity = movies_ranking.sort_values("popularity" , ascending=False)
plt.bar(popularity["original_title"].head(10) , popularity["vote_average"].head(10))
plt.xlabel("Movie Title")
plt.ylabel("Popularity")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_movies = scaler.fit_transform(movies_cleaned[["weighted_average" , "popularity"]])
normalized_movies = pd.DataFrame(scaled_movies , columns=["weighted_average" , "popularity"])
normalized_movies.head()
```

```python
movies_cleaned[["normalized_weight_average" , "normalized_popularity"]] = normalized_movies
movies_cleaned.head(2)
```

```python
movies_cleaned["score"] = (movies_cleaned["normalized_weight_average"] * 0.5) + (movies_cleaned["normalized_popularity"] * 0.5)
movies_scored = movies_cleaned.sort_values(["score"], ascending=False)
movies_scored[["original_title" , "normalized_weight_average" , "normalized_popularity" ,"score"]].head()
```
