```python
def convert_to_iob(labels):
    iob_labels = []
    for label in labels:
        if label['label']:
            entity_label = label['label'][0]
            iob_labels.append([(ent['start'], ent['end']+1, f'B-{entity_label}') for ent in label['points']])
    return iob_labels
```

```python
df['Y_iob'] = df['annotation'].apply(convert_to_iob)
```

```python
df['content'] = df['content'].apply(lambda x : x.lower())
```

```python
from keras.preprocessing.text import Tokenizer
y_tokenizer = Tokenizer()

flattened_labels = [label[0][2] for sublist in df['Y_iob'] for label in sublist]

y_tokenizer.fit_on_texts(flattened_labels)
```

```python
y_tokenizer.index_word
```

```python
flattened_labels
```

```python
classes = list(set(flattened_labels))
num_of_classes = len(classes)
num_of_classes
```

```python
y_tokenized = [y_tokenizer.texts_to_sequences([label[0][2]])[0] for sublist in df['Y_iob'] for label in sublist]
```

```python
y_tokenized
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
y_padded = pad_sequences(y_tokenized, padding='post', value=0 , maxlen=3015)
```

```python
y_padded
```

```python
max_len_y = max(len(x) for x in y_padded)
max_len_y
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['content'])
X_tokenized = tokenizer.texts_to_sequences(df['content'])
```

```python
tokenizer.index_word
```

```python
vocab_size = len(tokenizer.index_word) +1
vocab_size
```

```python
X_padded = pad_sequences(X_tokenized , padding='post')
X_padded
```

```python
max_len_x = max(len(x) for x in X_padded)
max_len_x
```

```python
X_padded.shape[1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense ,Bidirectional
```

```python
model = Sequential([
    Embedding(input_dim=vocab_size ,output_dim=100 ,input_length=max_len_x),
    Bidirectional(LSTM(300 ,return_sequences=True)),
    Bidirectional(LSTM(200 ,return_sequences=True)),
    Bidirectional(LSTM(100 ,return_sequences=True)),
    Dense(num_of_classes ,activation='softmax')
])

model.compile(loss='categorical_crossentropy' ,optimizer='adam' ,metrics=['accuracy'])
```

```python
model.summary()
```

```python
y_padded = [pad_sequences(seq, padding='post', dtype='object') for seq in df['Y_iob']]
y_padded
```

```python
import numpy as np
y_padded = np.array(y_padded)
```

```python
model.fit(X_padded ,y_padded ,epochs=5)
```
