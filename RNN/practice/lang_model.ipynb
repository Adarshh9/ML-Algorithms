```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from data4translator import data
```

```python
cleaned_data = re.sub(r'[^a-zA-Z\s]' , '' , data)
cleaned_data
```

```python
lowercase_data = cleaned_data.lower()
lowercase_data
```

```python
token = word_tokenize(lowercase_data)
token
```

```python
stop_words = set(stopwords.words('english'))
stop_words
```

```python
# filtered_data = [word for word in token if word.isalnum() and word not in stop_words]
filtered_data = [word for word in token if word.isalnum()]
filtered_data
```

```python
lengthof_filetered_data = len(filtered_data)
```

```python
sequence_length = 5

input_seq = []
output_seq = []

for i in range(len(filtered_data) - sequence_length):
    
    input = filtered_data[i : i+sequence_length]
    output = filtered_data[i+sequence_length]
    
    input_seq.append(input)
    output_seq.append(output)
    
```

```python
input_seq[0]
```

```python

output_seq[0]
```

```python
for i in range(5):
    print(f"Input Sequence: {input_seq[i]}, Output Sequence: {output_seq[i]}")
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts([filtered_data])
```

```python
tokenizer.word_index
```

```python
total_words = len(tokenizer.word_index)+1
```

```python
input_seq
```

```python
output_seq
```

```python
x_seq = tokenizer.texts_to_sequences(input_seq)
y_seq = tokenizer.texts_to_sequences(output_seq)
```

```python
y_seq
```

```python
y_seq = [item for sublist in y_seq for item in sublist]
```

```python
y_seq
```

```python
from keras.utils import to_categorical
y_seq = to_categorical(y_seq , num_classes=total_words)
```

```python
y_seq
```

```python
y_seq[0]
```

```python
len(x_seq)
```

```python
import numpy as np
x_seq = np.array(x_seq)
y_seq = np.array(y_seq)
```

LSTM MODEL

```python
from keras import Sequential
from keras.layers import Embedding ,LSTM ,GRU ,Bidirectional ,Attention ,Dense ,Dropout
from keras.callbacks import EarlyStopping
```

```python
model_lstm = Sequential([
    Embedding(input_dim=total_words , output_dim=200 , input_length=5),
    LSTM(300, return_sequences=True),
    Dropout(0.5),
    LSTM(150),
    Dense(units=total_words , activation="softmax")
])

model_lstm.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model_lstm.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=5 ,restore_best_weights=True)

model_lstm.fit(x_seq ,y_seq ,epochs=50 ,verbose=1 ,validation_split=0.2 ,batch_size=32)
```

```python
from keras.preprocessing.sequence import pad_sequences
import time
```

```python
text = "machine learning"

for i in range(10):
    tokenized_text = tokenizer.texts_to_sequences([text])[0]
    padded_token = pad_sequences(sequences=[tokenized_text] ,maxlen=5 ,padding='pre')
    pos = np.argmax(model_lstm.predict(padded_token))

    for word,index in tokenizer.word_index.items():
        if index == pos:
            text = text + " " + word
            print(text)
            time.sleep(1)
```
